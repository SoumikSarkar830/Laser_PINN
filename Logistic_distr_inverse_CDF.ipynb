```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0
s = 1

u = np.random.uniform(0,1,1000)

samples = mu + s*np.log(u/(1-u))

plt.hist(samples, bins = 50, density = True, alpha = 0.7, label = "Sampled data")
```

```python
# Plot the logistic distribution PDF

x = np.linspace(mu - 10*s, mu + 10*s, 1000)

pdf = np.exp(-(x-mu)/s)/(s*(1+np.exp(-(x-mu)/s))**2)

plt.plot(x,pdf,'r-',lw = 2, label = "True PDF")

plt.title("Sampling from Logistic Distribution using inverse CDF")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
