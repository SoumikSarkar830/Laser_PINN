```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0
b = 1

u = np.random.uniform(0,1,10000)

samples = np.where((u<0.5), mu + b*np.log(2*u),mu - b*np.log(2*(1-u)))

plt.hist(samples,bins = 50, density = True, alpha = 0.7, label = "Sampled data")
```

```python
# Plot for Laplace distribution PDF

x = np.linspace(mu-10*b, mu + 10*b, 1000)
pdf = (1/(2*b)) * np.exp((-np.abs(x-mu))/b)

plt.plot(x,pdf,'r-', lw =2, label = "True PDF")
plt.title("Laplace distribution sampling via inverse CDF")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
