```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

```python
alpha = 2
beta = 1.5
scale = 1/beta

u = np.random.uniform(0,1,10000)

samples = gamma.ppf(u,a=alpha,scale=scale)

plt.hist(samples, bins = 50, density = True, alpha = 0.7, label = "Sampled data")


```

```python
# Plot the Gamma Distribution PDF

x = np.linspace(0,np.max(samples),1000)
pdf = gamma.pdf(x,a=alpha,scale=scale)

plt.plot(x,pdf,'r-',lw = 2, label = "True PDF")

plt.title("Sampling from Gamma distribution using inverse CDF(Numerically)")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
