```python
import numpy as np
import matplotlib.pyplot as plt

!pip install --upgrade numpy scipy

from scipy.stats import norm
```

```python
mu = 0
sigma = 1

u = np.random.uniform(0,1,10000)

samples = norm.ppf(u,loc = mu,scale = sigma)

plt.hist(samples,bins = 50, density = True, alpha = 0.7, label = "Sampled data")
```

```python
# Plot the normal distribution PDF

x = np.linspace(mu-4*sigma, mu+4*sigma, 1000)
pdf = norm.pdf(x,loc = mu, scale = sigma)

plt.plot(x,pdf,'r-',lw = 2, label = "True PDF")

plt.title("Sampling from Normal distribution using inverse CDF (Numerically)")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.show()
```
