```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=2
b=5

u = np.random.uniform(0,1,10000)

samples = a + (b-a)*u

plt.hist(samples, bins = 50, density = True, alpha = 0.7, label = "Sampled data")
```

```python
# Uniform distribution plot for True PDF

x = np.linspace(a-1,b+1,1000)
pdf = np.where((x>=a) & (x<=b), 1/(b-a),0)

plt.plot(x,pdf,'r-', lw =2, label = "True PDF" )

plt.title("Uniform distribution sampling via inverse CDF")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
