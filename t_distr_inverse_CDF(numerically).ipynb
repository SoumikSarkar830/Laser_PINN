```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
```

```python
df = 5

u = np.random.uniform(0,1,10000)

samples = t.ppf(u,df)

plt.hist(samples, bins = 50, density = True, alpha = 0.7, label = "Sampled data")

```

```python
# Plot t distribution PDF

x = np.linspace(-10,10,1000)

pdf = t.pdf(x,df)

plt.plot(x,pdf,'r-',lw = 2, label = "True PDF")

plt.title("Sampling from t distribution using inverse CDF (Numerically)")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
