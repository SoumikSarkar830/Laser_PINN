```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda_ = 2

u = np.random.uniform(0,1,10000)

samples = -np.log(1-u) / lambda_

plt.hist(samples, bins = 50, density = True, alpha = 0.7, label = 'Sampled Data')   # Plotting the histogram for samples
```

```python
# Plotting True Exponential Distribution PDF for comparison

x = np.linspace(0, np.max(samples),1000)
pdf = lambda_ * np.exp(-lambda_*x)

plt.plot(x,pdf,'r-',lw = 2, label = 'True PDF')

plt.title("Exponential distribution sampling via inverse CDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

```
